```python
import cv2
from IPython.display import display, Image
import ipywidgets as widgets
from threading import Thread
import time

# Widget to display frames
img_widget = widgets.Image(format='jpeg')
display(img_widget)

cap = cv2.VideoCapture(0)  # 0 = default webcam

running = True

def stream():
    while running:
        ret, frame = cap.read()
        if not ret:
            break
        # Encode frame as JPEG
        _, buffer = cv2.imencode('.jpg', frame)
        img_widget.value = buffer.tobytes()
        time.sleep(0.03)  # ~30 FPS

# Start stream in background thread
t = Thread(target=stream)
t.start()
```

```python
import cv2
from ultralytics import YOLO

model = YOLO("yolo11n.pt")

cap = cv2.VideoCapture(0)  # en temp real for detect element or object
# 0 for one camera if have multi camera can add 0 1 2

while True:
    ret, frame = cap.read()   # take a picture
    # ret is return if can read return true or false
    # ret false frame = null
    if not ret:
        break

    h, w, _ = frame.shape   # size of photo for divide frame into three regions

    results = model(frame)

    for r in results:
        if len(r.boxes) == 0:
            continue

        x1, y1, x2, y2 = map(int, r.boxes.xyxy[0])
        cls_name = model.names[int(r.boxes.cls[0])]

        if cls_name == "cell phone":
            box_center = (x1 + x2) // 2
            area = (x2 - x1) * (y2 - y1)

            if box_center < w // 3:
                position = "LEFT"       # ✅ fixed typo: postion → position
            elif box_center > w * 2 // 3:
                position = "RIGHT"      # ✅ fixed typo: postion → position
            else:
                position = "CENTER"
                if area > 50000 :
                    position = "STOP"

            print("Cell phone position:", position)

            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(frame, cls_name + " " + position, (x1, y1 - 10),  # ✅ added position to label
                        cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)

    # Divide frame into 3 regions
    line1 = int(w / 3)
    line2 = int(2 * w / 3)

    cv2.line(frame, (line1, 0), (line1, h), (255, 0, 0), 2)
    cv2.line(frame, (line2, 0), (line2, h), (255, 0, 0), 2)

    # Show the frame
    cv2.imshow("Workshop - YOLO Detection", frame)

    # Press Q to quit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release everything cleanly
cap.release()
cv2.destroyAllWindows()
```
